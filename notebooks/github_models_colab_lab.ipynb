```python
from getpass import getpass

GITHUB_TOKEN = getpass("Paste your GitHub PAT here (input hidden): ")
print("Token stored securely.")
```

```python
from openai import OpenAI

client = OpenAI(
    base_url="https://models.github.ai/inference/v1",
    api_key=GITHUB_TOKEN
)

print("Client connected successfully.")
```

## First Model Call
This section demonstrates my first successful call to a GitHub Marketplace AI model using Python.


```python
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", "content": "Hello! Write one short fun fact about AI."}
    ]
)

print("First model call output:\n")
print(response.choices[0].message.content)
```

## Mini Project: Text Summarizer
This mini project takes a paragraph as input and summarizes it into exactly three bullet points using an AI model.


```python
# TEXT SUMMARIZER MINI PROJECT

text_to_summarize = """
E-commerce refers to buying and selling products and services online. It has changed how people shop by providing convenience and a wide variety of choices. Despite its benefits, issues such as online fraud and delayed deliveries remain challenges.
"""

prompt = f"""
Summarize the following text into EXACTLY 3 clear bullet points:

{text_to_summarize}
"""

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", "content": prompt}
    ]
)

summary = response.choices[0].message.content

print("ORIGINAL TEXT:\n")
print(text_to_summarize)

print("\nSUMMARY (3 Bullet Points):\n")
print(summary)
```

```python
with open("summary_output.txt", "w") as file:
    file.write(summary)

print("Summary saved to summary_output.txt")
```
